```python
#import essential libraries

import pandas as pd

#load the superstore dataset

df = pd.read_csv('Sample - Superstore.csv', encoding='latin1')

#display the first 5 rows of the data

df.head()
```

```python
#checking data info (number of rows, columns, missing values)
df.info()

#checking for missing values
print("\nMissing values per column:")
print(df.isnull().sum())

#converting 'Order Date' and 'Ship Date' to datetime
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], errors='coerce')

#create a new column: Profit Margin = Profit / Sales
df['Profit Margin'] = df['Profit'] / df['Sales']

#creating discount categories
df['Discount Band'] = pd.cut(df['Discount'],
                             bins=[-1, 0.1, 0.2, 1],
                             labels=['Low (<=10%)', 'Medium (10-20%)', 'High (>20%)'])

#Preview cleaned data
df.head()
```

```python
#Save cleaned data to a new CSV file
df.to_csv('cleaned_superstore.csv', index=False)
print("Cleaned data saved as 'cleaned_superstore.csv'")

#Show first few rows again
df.head()
```

```python
#Save the cleaned data with UTF-8 encoding
df.to_csv('cleaned_superstore.csv', index=False, encoding='utf-8')
print("File saved with UTF-8 encoding — special characters will now appear correctly!")
```
